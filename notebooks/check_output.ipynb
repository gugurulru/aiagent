```python
from dotenv import load_dotenv
import os

load_dotenv()

from state_schema import create_initial_state
from agents import EthicsCriteriaGenerator

# State ÏÉùÏÑ±
state = create_initial_state("Vara", "medical")

# Agent Ïã§Ìñâ
generator = EthicsCriteriaGenerator(openai_api_key=os.getenv("OPENAI_API_KEY"))
state = generator.execute(state)

# Í≤∞Í≥º ÌôïÏù∏
criteria = state["ethics_evaluation_criteria"]
for cat_key, cat_data in criteria['categories'].items():
    print(f"\n{cat_data['category']}: {len(cat_data['criteria'])}Í∞ú Í∏∞Ï§Ä")
    for i, crit in enumerate(cat_data['criteria'], 1):
        print(f"  {i}. {crit['title']}")
```

```python
from dotenv import load_dotenv
import os
load_dotenv()
TAVILY_API_KEY = os.getenv("TAVILY_API_KEY")
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

from state_schema import create_initial_state
from agents import WebCollectionAgent, SpecializedCollectionAgent
from agents.ethics_criteria_generator import EthicsCriteriaGenerator
from agents.ethics_evaluator import EthicsEvaluator

# State ÏÉùÏÑ± (Cell 2)
state = create_initial_state("Vara", "medical")

# ÏàòÏßë (Cell 3-4)
web_agent = WebCollectionAgent(
    tavily_api_key=os.getenv("TAVILY_API_KEY"),
    openai_api_key=os.getenv("OPENAI_API_KEY"))
state = web_agent.execute(state)

spec_agent = SpecializedCollectionAgent(
    tavily_api_key=os.getenv("TAVILY_API_KEY"),
    openai_api_key=os.getenv("OPENAI_API_KEY")
)
state = spec_agent.execute(state)


# ÌèâÍ∞Ä Í∏∞Ï§Ä ÏÉùÏÑ± (Cell 5)
generator = EthicsCriteriaGenerator(openai_api_key=os.getenv("OPENAI_API_KEY"))
state = generator.execute(state)

# ÌèâÍ∞Ä Ïã§Ìñâ (Cell 6)
evaluator = EthicsEvaluator(openai_api_key=OPENAI_API_KEY)
state = evaluator.execute(state)

# Í≤∞Í≥º ÌôïÏù∏ (Cell 7)
print(f"Ï†êÏàò: {state['ethics_score']['total']}/100")
```
